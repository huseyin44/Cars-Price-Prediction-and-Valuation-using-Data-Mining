```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

#verileri ekleme ve diğer veriambarıyla bağlantısız olan sutunlari çıkarma
CarAdVerileri = pd.read_csv('car_ad.csv',encoding = "ISO-8859-1")

CarAdVerileri.drop(columns='body', axis = 1,inplace=True)
CarAdVerileri.drop(columns='registration', axis = 1,inplace=True)
CarAdVerileri.drop(columns='drive', axis = 1,inplace=True)
CarAdVerileri.drop(columns='model', axis = 1,inplace=True)

#sadece Audi arabalarini filtreleme
AudiFiltre=CarAdVerileri["car"].isin(['Audi']) 
AudiFiltre=CarAdVerileri[AudiFiltre]
#nan değeri olan satırları siliyior
AudiFiltre.dropna(how="any",inplace=True) 
#verideki sütünlarin yerini değiştirme
columns_titles = ["car","year","mileage","engV","engType","price"]
AudiFiltre=AudiFiltre.reindex(columns=columns_titles)

#sadece Skoda arabalarini filtreleme
SkodaFiltre=CarAdVerileri["car"].isin(['Skoda']) 
SkodaFiltre=CarAdVerileri[SkodaFiltre]
#verideki sütünlarin yerini değiştirme
SkodaFiltre=SkodaFiltre.reindex(columns=columns_titles)
#nan değeri olan satırları siliyior
SkodaFiltre.dropna(how="any",inplace=True) 

#sadece Bmw arabalarini filtreleme
BmwFiltre=CarAdVerileri["car"].isin(['BMW']) 
BmwFiltre=CarAdVerileri[BmwFiltre]
#verideki sütünlarin yerini değiştirme
BmwFiltre=BmwFiltre.reindex(columns=columns_titles)
#nan değeri olan satırları siliyior
BmwFiltre.dropna(how="any",inplace=True) 

#sadece Toyota arabalarini filtreleme
ToyotaFiltre=CarAdVerileri["car"].isin(['Toyota']) 
ToyotaFiltre=CarAdVerileri[ToyotaFiltre]
#verideki sütünlarin yerini değiştirme
ToyotaFiltre=ToyotaFiltre.reindex(columns=columns_titles)
#nan değeri olan satırları siliyior
ToyotaFiltre.dropna(how="any",inplace=True) 

#mili km cinsine çevirme
AudiFiltre.loc[:,'mileage'] *=1.609344
SkodaFiltre.loc[:,'mileage'] *=1.609344
BmwFiltre.loc[:,'mileage'] *=1.609344
ToyotaFiltre.loc[:,'mileage'] *=1.609344

AudiFiltre.head()
```

```python
#Diğer veriler içinde aynı işlemler
AudiVerileri = pd.read_csv('audi.csv')
BmwVerileri = pd.read_csv('bmw.csv')
SkodaVerileri = pd.read_csv('skoda.csv')
ToyotaVerileri = pd.read_csv('toyota.csv')

AudiVerileri = AudiVerileri.drop('mpg', axis = 1)
AudiVerileri = AudiVerileri.drop('tax', axis = 1)
AudiVerileri = AudiVerileri.drop('transmission', axis = 1)
AudiVerileri = AudiVerileri.drop('model', axis = 1)
#araba markasını ekleme
AudiVerileri.insert(0,column="Marka",value="Audi") 
#Price Veri tipini int ten Float a çevirme
AudiVerileri['price'] = AudiVerileri['price'].astype('float')
 
SkodaVerileri = SkodaVerileri.drop('mpg', axis = 1)
SkodaVerileri = SkodaVerileri.drop('tax', axis = 1)
SkodaVerileri = SkodaVerileri.drop('transmission', axis = 1)
SkodaVerileri = SkodaVerileri.drop('model', axis = 1)
#araba markasını ekleme
SkodaVerileri.insert(0,column="Marka",value="Skoda") 
#Price Veri tipini int ten Float a çevirme
SkodaVerileri['price'] = SkodaVerileri['price'].astype('float')
 
BmwVerileri = BmwVerileri.drop('mpg', axis = 1)
BmwVerileri = BmwVerileri.drop('tax', axis = 1)
BmwVerileri = BmwVerileri.drop('transmission', axis = 1)
BmwVerileri = BmwVerileri.drop('model', axis = 1)
#araba markasını ekleme
BmwVerileri.insert(0,column="Marka",value="BMW")  
#Price Veri tipini int ten Float a çevirme
BmwVerileri['price'] = BmwVerileri['price'].astype('float')
 
ToyotaVerileri = ToyotaVerileri.drop('mpg', axis = 1)
ToyotaVerileri = ToyotaVerileri.drop('tax', axis = 1)
ToyotaVerileri = ToyotaVerileri.drop('transmission', axis = 1)
ToyotaVerileri = ToyotaVerileri.drop('model', axis = 1)
#araba markasını ekleme
ToyotaVerileri.insert(0,column="Marka",value="Toyota")  
#Price Veri tipini int ten Float a çevirme
ToyotaVerileri['price'] = ToyotaVerileri['price'].astype('float')
 
#Sütünlari Sıralama
columns_titles = ["Marka","year","mileage","engineSize","fuelType","price"]
AudiVerileri=AudiVerileri.reindex(columns=columns_titles)
SkodaVerileri=AudiVerileri.reindex(columns=columns_titles)
ToyotaVerileri=AudiVerileri.reindex(columns=columns_titles)
BmwVerileri=AudiVerileri.reindex(columns=columns_titles)

#AudiVerileri,SkodaVerileri,ToyotaVerileri,BmwVerileri 'verilerindeki PRİCE EURO tipinde
#AudiFiltre,SkodaFiltre,ToyotaFiltre,BmwFiltre 'verilerindeki PRİCE USD tipinde
# USD=0,91*Euro bu yüzden AudiVerileri,SkodaVerileri,ToyotaVerileri,BmwVerileri 'verilerinin price ini 0,91 ile çarptım
AudiVerileri.loc[:,'price'] *= 0.91
SkodaVerileri.loc[:,'price'] *= 0.91
ToyotaVerileri.loc[:,'price'] *= 0.91
BmwVerileri.loc[:,'price'] *= 0.91

SkodaVerileri.head()
```

```python
# 3. Veri Seti
Polonya_Audi = pd.read_csv('Car_Prices_Poland_Kaggle.csv',encoding = "ISO-8859-1")

Polonya_Audi.drop(columns='model', axis = 1,inplace=True)
Polonya_Audi.drop(columns='generation_name', axis = 1,inplace=True)
Polonya_Audi.drop(columns='city', axis = 1,inplace=True)
Polonya_Audi.drop(columns='Unnamed: 0', axis = 1,inplace=True)
Polonya_Audi.drop(columns='province', axis = 1,inplace=True)

#sadece audi marka araçları filtrele
Polonya_Audi_Filtre=Polonya_Audi["mark"].isin(['audi']) 
Polonya_Audi_Filtre=Polonya_Audi[Polonya_Audi_Filtre]

#Sutunlardaki Audi ismini düzeltme
Polonya_Audi_Filtre['mark'] = Polonya_Audi_Filtre['mark'].map({'audi': "Audi",})

#sutun yerlerini ve ismini degiştirme
columns_titles = ["mark","year","mileage","fuel","vol_engine","price"]

#DataTiplerini Dönüştürme
Polonya_Audi_Filtre['price'] = Polonya_Audi_Filtre['price'].astype('float')
Polonya_Audi_Filtre['vol_engine'] = Polonya_Audi_Filtre['vol_engine'].astype('float')

#1 EUR => 4,70 PLN fiyati Polonya parasından Euroya çevirme ve yuvarlama
Polonya_Audi_Filtre.loc[:,'price'] =round((Polonya_Audi_Filtre.loc[:,'price'] / 4.70),0)
#Vol_engine diğer engineSize lara eşitleme
Polonya_Audi_Filtre.loc[:,'vol_engine'] =round((Polonya_Audi_Filtre.loc[:,'vol_engine'] / 1000),0)

Polonya_Audi_Filtre
```

```python
#Sütünlari Yeniden İsimlendirme
YeniSutunIsım={"Marka": "Marka","year": "ÜretimYılı","mileage":"KM", "engineSize": "MotorHacmi","fuelType": "YakıtTipi","price": "Fiyatı"}
YeniSutunIsım2={"car": "Marka","year": "ÜretimYılı","mileage":"KM", "engV": "MotorHacmi","engType": "YakıtTipi","price": "Fiyatı"}
YeniSutunIsım3={"mark": "Marka","year": "ÜretimYılı","mileage":"KM", "vol_engine": "MotorHacmi","fuel": "YakıtTipi","price": "Fiyatı"}

BmwVerileri.rename(YeniSutunIsım, axis="columns", inplace=True)
AudiVerileri.rename(YeniSutunIsım, axis="columns", inplace=True)
ToyotaVerileri.rename(YeniSutunIsım, axis="columns", inplace=True)
SkodaVerileri.rename(YeniSutunIsım, axis="columns", inplace=True)

BmwFiltre.rename(YeniSutunIsım2, axis="columns", inplace=True)
SkodaFiltre.rename(YeniSutunIsım2, axis="columns", inplace=True)
ToyotaFiltre.rename(YeniSutunIsım2, axis="columns", inplace=True)
AudiFiltre.rename(YeniSutunIsım2, axis="columns", inplace=True)

Polonya_Audi_Filtre.rename(YeniSutunIsım3, axis="columns", inplace=True)
```

```python
TumVeriler=pd.concat([BmwFiltre, SkodaFiltre],axis = 0)
TumVeriler=pd.concat([TumVeriler, ToyotaFiltre],axis = 0)
TumVeriler=pd.concat([TumVeriler, AudiFiltre],axis = 0)
TumVeriler=pd.concat([TumVeriler, SkodaVerileri],axis = 0)
TumVeriler=pd.concat([TumVeriler, AudiVerileri],axis = 0)
TumVeriler=pd.concat([TumVeriler, ToyotaVerileri],axis = 0)
TumVeriler=pd.concat([TumVeriler, BmwVerileri],axis = 0)
TumVeriler=pd.concat([TumVeriler, Polonya_Audi_Filtre],axis = 0)
#verileri görme
#TumVeriler.iloc[[1000,1,20000,11111]]

TumVeriler.head(5)
```

```python
#yakıt tiplerini görme
TumVeriler.groupby("YakıtTipi").mean()
```

```python
#araba markalarını görme
TumVeriler.groupby("Marka").mean()
```

```python
#Kategorik Verileri sayısal veriye çevirme ve veri setini tekrar düzenleme
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
ohe=OneHotEncoder()

Marka=TumVeriler.iloc[:,0:1].values
Marka[:,0]=le.fit_transform(Marka[:,0])  
Marka=ohe.fit_transform(Marka).toarray()
Marka=pd.DataFrame(data=Marka,index=range(56688),columns=['Audi','BMW','Skoda','Toyota'])

UretimYili=TumVeriler.iloc[:,1:2].values
UretimYili=pd.DataFrame(data=UretimYili,index=range(56688),columns=['UretimYili'])

KM=TumVeriler.iloc[:,2:3].values
KM=pd.DataFrame(data=KM,index=range(56688),columns=['KM'])

MotorHacmi=TumVeriler.iloc[:,3:4].values
MotorHacmi=pd.DataFrame(data=MotorHacmi,index=range(56688),columns=['MotorHacmi'])

YakitTipi=TumVeriler.iloc[:,4:5].values
YakitTipi=ohe.fit_transform(YakitTipi).toarray()
YakitTipi=pd.DataFrame(data=YakitTipi,index=range(56688),columns=['CNG','Diesel','Electric','Gas','Gasoline'
                                                                  ,'Hybrid','LPG','Other','Petrol'])
Fiyati=TumVeriler.iloc[:,5:6].values
Fiyati=pd.DataFrame(data=Fiyati,index=range(56688),columns=['Fiyati'])

KategorizeEdilmisVeriler=pd.concat([Marka,UretimYili],axis=1)
KategorizeEdilmisVeriler=pd.concat([KategorizeEdilmisVeriler,KM],axis=1)
KategorizeEdilmisVeriler=pd.concat([KategorizeEdilmisVeriler,MotorHacmi],axis=1)
KategorizeEdilmisVeriler=pd.concat([KategorizeEdilmisVeriler,YakitTipi],axis=1)
KategorizeEdilmisVeriler=pd.concat([KategorizeEdilmisVeriler,Fiyati],axis=1)

KategorizeEdilmisVeriler
```

```python
#eksik verileri görme
KategorizeEdilmisVeriler.isnull().sum()
```

```python
#veriyi inceleme
KategorizeEdilmisVeriler.describe()
```

```python
#motorhacmi ve fiyatı  0 olamacığı için onları çıkartıyıoruz
KategorizeEdilmisVeriler=KategorizeEdilmisVeriler[KategorizeEdilmisVeriler.MotorHacmi !=0]
KategorizeEdilmisVeriler=KategorizeEdilmisVeriler[KategorizeEdilmisVeriler.Fiyati !=0]

KategorizeEdilmisVeriler.describe()
```
